```python
# Para bajar datos de prepara, solo me quedo con los que compran al menos 2 veces en el año.
# Pruebo con > 1, si aun son muchos datos para el posterior calculo donde me dejaba con memory out, paso a > 2 y 
# sigo probando.
# Me dio en mi maquina 'MemoryError', paso a 2, paso a 3 y menor de 100 para sacar los 2 casos atipicos

df_prepara_reco1 = pd.DataFrame(df_prepara.Documento.value_counts())
df_prepara_reco1['index1'] = df_prepara_reco1.index
df_prepara_reco1.rename(columns={"Documento": "repeticion", "index1": "Documento"}, inplace=True)
a = (df_prepara_reco1.repeticion > 3) & (df_prepara_reco1.repeticion < 100)
a.sum()
```

```python
a
```

```python
# Lista de documentos con repeticiones de compra
df_prepara_reco2 = np.array(df_prepara_reco1[a]['Documento'])
df_prepara_reco2.sort()
df_prepara_reco2
```

```python
# Ahora tengo reducido el numero de clientes a los que repiten compras para chequear preferidos
df_prepara_reco3 = df_prepara.loc[(df_prepara['Documento'].isin(df_prepara_reco2))]
df_prepara_reco3.sample(10)
```

```python
# Armo df para exportar por ID
df_prepara_recomendacion = df_prepara_reco3.reindex(columns = ['Documento', 'ProductId','Quantity'])
df_prepara_recomendacion.rename(columns={'Documento':'customerId','ProductId':'productId', \
                                         'Quantity':'purchase_count'}, inplace=True)

df_prepara_recomendacion.to_csv('Data.csv')

df_prepara_recomendacion
```

```python
#### Se preparan datos para correr recomendaciones con un rating figurado tanto para datos Id como para Typo
```

```python
df_prepara_recomendacion.purchase_count[a] = 4
df_prepara_recomendacion.purchase_count[b] = 5
```

## Preparamos datos para otra recomendacion de a prueba

#### Tenemos que armar una reducida de data para el engine sistem recomendation que no soporta una gran base

```python
# Achicamos con la data
a = df_prepara_reco3.OrderDate > '2019-04-15'
```

```python
a.sum()
```

```python
# Armo df para exportar por ID
df_prepara_recoreducida = df_prepara_reco3[a].reindex(columns = ['Documento', 'ProductId','Quantity', 'Id', 'ProductType'])
df_prepara_recoreducida.rename(columns={'Documento':'userId','Id':'repoId', 'Quantity':'rating', 'ProductId':'categories',\
                                       'ProductType':'title'}, inplace=True)
```

```python
df_prepara_recoreducida['repoId'] = df_prepara_recoreducida['title'].copy()
```

```python
df_prepara_recoreducida.sample(3)
```

```python
#### Se preparan datos para correr recomendaciones con un rating figurado tanto para datos Id como para Typo
```

```python
a = (df_prepara_recoreducida.rating > 3) & (df_prepara_recoreducida.rating < 10)
b = (df_prepara_recoreducida.rating > 9)
```

```python
df_prepara_recoreducida.rating[a] = 4
df_prepara_recoreducida.rating[b] = 5
```

```python
df_prepara_recoreducida.rating.unique()
```

```python
# para sacar el unnamed:0, hay que poner index=False... despues de tantos notebookssssss

df_prepara_recoreducida[{'userId', 'title', 'rating'}].to_csv('Datarecoreducida.csv', index=False)

df_prepara_recoreducida.sample(5)
```
