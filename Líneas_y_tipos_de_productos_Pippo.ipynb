```python
#Clasificación de Productos:

#Patrones por Tipo de Producto:

#Baño
tp_b_toallas = r"(Toalla|Toallón)"
tp_b_jabones = r"(Jabón|Jabonera|Jabones)"
tp_b_accesorios = r"(Baño|Portacepillo de Dientes|Dientes|Alfombra de Baño)"

#Dormitorio
tp_d_sabanas = r"(Sábana|Sabana|Sábanas)"
tp_d_cama = r"(Sommier|Cubresommier|Cubrecolchón|Cama)"
tp_d_acolchados = r"(Acolchado|Plumón|Frazada|Manta|Cubrecama|Cubrema)"
tp_d_bebe = r"(Nido Bebé|Chichonera)"
tp_d_almoh = r"(Almohada)"

#Cocina
tp_c_vajilla = r"(Manopla|Cubiertos|Vajilla)"
tp_c_manteleria = r"(Mantel|Repasadores|Repasador|Individual|Set de Servilletas)"
tp_c_accesorios   = r"(Secaplatos|Secador de Vajilla|Servilletero)"

#Accesorios
tp_a_almohadones  = r"(Funda|Fundas|Almohadón|Relleno)"
tp_a_velas  = r"(Vela|Vela Pajarito|Fanal)"
tp_a_alfombras  = r"(Alfombra de Ambiente)"
tp_a_aromatizadores  = r"(Aromatizador|Difusor|Aromática|Aromatica|Aromatizante|Dosificador)"
tp_a_muebles  = r"(Tablero|Contenedores|Cooler|Carro|Adorno|Mat de Secado)"

#Aire Libre
tp_p_sillas  = r"(Reposera|Playa)"
tp_p_lonas  = r"(Lona|Playera)"

#Organizadores
tp_o_cajs  = r"(Cajón|Cajon|Caja|Cajas)"
tp_o_placard  = r"(Perchas|Perchero|Bolsa|Organizador|Canasta)"
tp_o_guardado  = r"(Cesto|Contenedor|Banco|Bandejas)"

#Decoración
tp_m_cortinas  = r"(Cortina|Cortinas)"
tp_m_muebles  = r"(Poltrona|Banquito|Mesa|Silla|Puff|Camino|Tatami|Jardín Esquinero|Jardín Simple)"
tp_m_bandejas  = r"(Bandeja|Banjeda|Bandejas|Banjedas)"
tp_m_adornos  = r"(Marco|Cuadro|Espejo|Florero|Algodonera|Revistero|Portarretratos|Portarretrato|Portafotos|Portavela|Set de Botellas|Posavasos)"
tp_m_jardin  = r"(Jardín Vertical|Maceta)"

#Ropa
tp_r_dormir  = r"(Pijama|Remera|Camisón|Camison|Bata)"
tp_r_calle  = r"(Mochila|Saquito|Musculosa|Poncho|Túnica|Tunica)"
tp_r_trabajo  = r"(Delantal)"

#Patrones por Linea de Producto:

lp_baño = tp_b_toallas + "|" + tp_b_jabones + "|" + tp_b_accesorios
lp_dormitorio = tp_d_sabanas + "|" + tp_d_cama  + "|" + tp_d_acolchados  + "|" + tp_d_bebe  + "|" + tp_d_almoh
lp_cocina =  tp_c_vajilla  + "|" + tp_c_manteleria  + "|" + tp_c_accesorios
lp_accesorios = tp_a_almohadones  + "|" + tp_a_velas + "|" + tp_a_alfombras + "|" + tp_a_aromatizadores + "|" + tp_a_muebles
lp_airelibre =  tp_p_sillas + "|" + tp_p_lonas
lp_organizadores = tp_o_cajs + "|" + tp_o_placard + "|" + tp_o_guardado
lp_deco =  tp_m_cortinas + "|" + tp_m_muebles  + "|" + tp_m_bandejas  + "|" + tp_m_adornos + "|" + tp_m_jardin
lp_ropa = tp_r_dormir  + "|" + tp_r_calle + "|" + tp_r_trabajo


#Barrido del campo 'Name' para aplicar las clasificaciones mediante expresiones regulares:
b = df[df['Name'].str.contains(lp_baño, flags=re.I, regex=True)]
b1 = df[df['Name'].str.contains(tp_b_toallas, flags=re.I, regex=True)]
b2 = df[df['Name'].str.contains(tp_b_jabones, flags=re.I, regex=True)]
b3 = df[df['Name'].str.contains(tp_b_accesorios, flags=re.I, regex=True)]

d = df[df['Name'].str.contains(lp_dormitorio, flags=re.I, regex=True)]
d1 = df[df['Name'].str.contains(tp_d_sabanas, flags=re.I, regex=True)]
d2 = df[df['Name'].str.contains(tp_d_cama, flags=re.I, regex=True)]
d3 = df[df['Name'].str.contains(tp_d_acolchados, flags=re.I, regex=True)]
d4 = df[df['Name'].str.contains(tp_d_bebe, flags=re.I, regex=True)]
d5 = df[df['Name'].str.contains(tp_d_almoh, flags=re.I, regex=True)]

c = df[df['Name'].str.contains(lp_cocina, flags=re.I, regex=True)]
c1 = df[df['Name'].str.contains(tp_c_vajilla, flags=re.I, regex=True)]
c2 = df[df['Name'].str.contains(tp_c_manteleria, flags=re.I, regex=True)]
c3 = df[df['Name'].str.contains(tp_c_accesorios, flags=re.I, regex=True)]

a = df[df['Name'].str.contains(lp_accesorios, flags=re.I, regex=True)]
a1 = df[df['Name'].str.contains(tp_a_almohadones, flags=re.I, regex=True)]
a2 = df[df['Name'].str.contains(tp_a_velas, flags=re.I, regex=True)]
a3 = df[df['Name'].str.contains(tp_a_alfombras, flags=re.I, regex=True)]
a4 = df[df['Name'].str.contains(tp_a_aromatizadores, flags=re.I, regex=True)]
a5 = df[df['Name'].str.contains(tp_a_muebles, flags=re.I, regex=True)]

p = df[df['Name'].str.contains(lp_airelibre, flags=re.I, regex=True)]
p1 = df[df['Name'].str.contains(tp_p_sillas, flags=re.I, regex=True)]
p2 = df[df['Name'].str.contains(tp_p_lonas, flags=re.I, regex=True)]

o = df[df['Name'].str.contains(lp_organizadores, flags=re.I, regex=True)]
o1 = df[df['Name'].str.contains(tp_o_cajs, flags=re.I, regex=True)]
o2 = df[df['Name'].str.contains(tp_o_placard, flags=re.I, regex=True)]
o3 = df[df['Name'].str.contains(tp_o_guardado, flags=re.I, regex=True)]

m = df[df['Name'].str.contains(lp_deco, flags=re.I, regex=True)]
m1 = df[df['Name'].str.contains(tp_m_cortinas, flags=re.I, regex=True)]
m2 = df[df['Name'].str.contains(tp_m_muebles, flags=re.I, regex=True)]
m3 = df[df['Name'].str.contains(tp_m_bandejas, flags=re.I, regex=True)]
m4 = df[df['Name'].str.contains(tp_m_adornos, flags=re.I, regex=True)]
m5 = df[df['Name'].str.contains(tp_m_jardin, flags=re.I, regex=True)]

r = df[df['Name'].str.contains(lp_ropa, flags=re.I, regex=True)]
r1 = df[df['Name'].str.contains(tp_r_dormir, flags=re.I, regex=True)]
r2 = df[df['Name'].str.contains(tp_r_calle, flags=re.I, regex=True)]
r3 = df[df['Name'].str.contains(tp_r_trabajo, flags=re.I, regex=True)]

#Indexado de las listas
index_b = list(b.index)
index_b1 = list(b1.index)
index_b2 = list(b2.index)
index_b3 = list(b3.index)
index_d = list(d.index)

index_d1 = list(d1.index)
index_d2 = list(d2.index)
index_d3 = list(d3.index)
index_d4 = list(d4.index)
index_d5 = list(d5.index)

index_c = list(c.index)
index_c1 = list(c1.index)
index_c2 = list(c2.index)
index_c3 = list(c3.index)

index_a = list(a.index)
index_a1 = list(a1.index)
index_a2 = list(a2.index)
index_a3 = list(a3.index)
index_a4 = list(a4.index)
index_a5 = list(a5.index)

index_p = list(p.index)
index_p1 = list(p1.index)
index_p2 = list(p2.index)

index_o = list(o.index)
index_o1 = list(o1.index)
index_o2 = list(o2.index)
index_o3 = list(o3.index)

index_m = list(m.index)
index_m1 = list(m1.index)
index_m2 = list(m2.index)
index_m3 = list(m3.index)
index_m4 = list(m4.index)
index_m5 = list(m5.index)

index_r = list(r.index)
index_r1 = list(r1.index)
index_r2 = list(r2.index)
index_r3 = list(r3.index)


#Aplicación de indización para completar tipos y lineas de producto:
df.loc[index_b,'ProductLine'] = "Baño"
df.loc[index_b1,'ProductType'] = "Toallas y Toallones"
df.loc[index_b2,'ProductType'] = "Jabones y Jaboneras"
df.loc[index_b3,'ProductType'] = "Accesorios"

df.loc[index_d,'ProductLine'] = "Dormitorio"
df.loc[index_d1,'ProductType'] = "Sábanas"
df.loc[index_d2,'ProductType'] = "Cama"
df.loc[index_d3,'ProductType'] = "Acolchados y Frazadas"
df.loc[index_d4,'ProductType'] = "Bebé"
df.loc[index_d5,'ProductType'] = "Almohadas"

df.loc[index_c,'ProductLine'] = "Cocina"
df.loc[index_c1,'ProductType'] = "Vajilla"
df.loc[index_c2,'ProductType'] = "Mantelería"
df.loc[index_c3,'ProductType'] = "Accesorios"

df.loc[index_a,'ProductLine'] = "Accesorios"
df.loc[index_a1,'ProductType'] = "Almohadones"
df.loc[index_a2,'ProductType'] = "Velas"
df.loc[index_a3,'ProductType'] = "Alfombras"
df.loc[index_a4,'ProductType'] = "Aromatizadores"
df.loc[index_a5,'ProductType'] = "Muebles"

df.loc[index_p,'ProductLine'] = "Aire Libre"
df.loc[index_p1,'ProductType'] = "Sillas"
df.loc[index_p2,'ProductType'] = "Lonas"

df.loc[index_o,'ProductLine'] = "Organizadores"
df.loc[index_o1,'ProductType'] = "Cajas y Cajones"
df.loc[index_o2,'ProductType'] = "Placard"
df.loc[index_o3,'ProductType'] = "Guardado"

df.loc[index_m,'ProductLine'] = "Decoración"
df.loc[index_m1,'ProductType'] = "Cortinas"
df.loc[index_m2,'ProductType'] = "Muebles"
df.loc[index_m3,'ProductType'] = "Bandejas"
df.loc[index_m4,'ProductType'] = "Adornos"
df.loc[index_m5,'ProductType'] = "Jardín"

df.loc[index_r,'ProductLine'] = "Ropa"
df.loc[index_r1,'ProductType'] = "Para Dormir"
df.loc[index_r2,'ProductType'] = "Para Uso"
df.loc[index_r3,'ProductType'] = "Para Trabajo"
```
